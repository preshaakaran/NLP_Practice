```python
import nltk
from nltk import word_tokenize
from nltk.corpus import treebank
import nltk
nltk.download('treebank')
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Train the HMM POS Tagger on the Treebank corpus
train_data = treebank.tagged_sents()[:3000]
hmm_tagger = nltk.HiddenMarkovModelTagger.train(train_data)
```

```python
# POS tagging function using the trained HMM model
def pos_tag(sentence):
    words = word_tokenize(sentence)
    tags = hmm_tagger.tag(words)
    return tags
```

```python
# Example usage
sentence = "Narayana Murthy and Akshata Murty visited iconic Bengaluru ice cream joint."
tags = pos_tag(sentence)
print(tags)
```
